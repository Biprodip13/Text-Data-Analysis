```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
videos=pd.read_csv('F:\Plots\Project 2 dataset/Indiacomments.csv',error_bad_lines=False)
videos.head()
```

# Problem Statement 1: Performing Sentiment analysis on Youtube Comments

```python
from textblob import TextBlob
```

```python
TextBlob('I have been following you from the start of your...').sentiment.polarity
# 1=Poitive polarity
# 0=Neutral Polarity
#-1=Negative Polarity
```

```python
polarity_store=[]
for i in videos['comment_text']:
    try:
        polarity_store.append(TextBlob(i).sentiment.polarity)
    except:
        polarity_store.append(0)
```

```python
videos['polarity_store']=polarity_store
```

```python
videos.head(10)
```

# Problem Statement 2: Performing EDA analysis for positive Comments

```python
comments_positive=videos[videos['polarity_store']==1]
comments_positive.head(10)
```

```python
from wordcloud import WordCloud, STOPWORDS
total_comment=(''.join(comments_positive['comment_text']))
len(total_comment)
wordcloud = WordCloud(width=1000, height=500, stopwords=set(STOPWORDS)).generate(total_comment)
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

# Problem Statement 3: Performing EDA analysis for Negative Comments
